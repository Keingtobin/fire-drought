## Fire and Drought
If you live in the North American West you are probably very familiar with one of it's most common natural disasters, wildfires. Wildfire frequency and intensity has been increasing with the worsening conditions of global warming. While there are many causes for wildfires, ranging from intentional arson to a downed powerline, global warming has made the fires from these events worse. This is in part due to worsening drought conditions caused by the warming temperatures. When air temperatures rise, the air is able to hold a larger amount of moisture. This especially affects areas like the West, because the already dry forests are losing more and more moisture to the atmoshphere. With drought frequency increasing, high fire risk weather conditions are more likely to line up with drought conditions. This leads to compounding the risk of wildfires. [[1]](https://doi.org/10.1038/s41612-022-00248-4)

### California 2020
In the year 2020, California saw its worst fire year on yecord. During this year alone, wildfires killed over 30 people and burned over 4 million acres. [[2]](https://www.ebsco.com/research-starters/history/2020-california-wildfires) While many things likely led to this being the worst wildfire year for the state on record, I am going to look at the correlation between drought conditions and the wildfires in order to see how drought affected the year.

2020 did not begin with the entire state in drought. There were some record dry areas in central and northern California [[3]](https://www.drought.gov/drought-status-updates/drought-status-update-and-2020-recap-california-nevada), but in general the rest of the state was not seeing drought. However, as the year progressed more of the state was considered in a drought. This issue started getting worse with record high temperatures in August [[3]](https://www.drought.gov/drought-status-updates/drought-status-update-and-2020-recap-california-nevada). 

## Methods
In order to analyze how drought may have played a part in the 2020 California wildfire season, I am going to be using publicly available datasets. The first dataset is from [MTBS](https://www.mtbs.gov/). MTBS tracks wildfire events in the United States, logging fires above 1,000 acres in the West and 500 acres in the East. With this data, I am able to grab all of Californias wildfires that were ignited in 2020 that burned 1,000 acres and over. The next dataset I am using is from the [U.S. Drought Monitor](https://droughtmonitor.unl.edu/). They track all sorts of data related to drought and compile it into an index, with D0 being abnormally dry conditions, ranging all the way to D4-exceptional drought. Specifically, I am going to be using their GIS data. This data provides us with shapefiles of all the drought conditions in the United States, updated weekly.

Everything else in this methods section is going to be code. If you have no interest in the nitty gritty of how the analysis was run, skip to the bottom for my results/discussion.

### Step 0: Import Needed Packages

```python
# Packages for working with filepath and operating system
import os
import pathlib
import glob
# Packages for downloading the data (that we can get programatically)
import requests
from zipfile import ZipFile
import osmnx as osm # For working with OpenStreetMap
import datetime # Useful for extracting a datetime object out of filenames
# Packages for working with spatial and tabular data
import geopandas as gpd
import pandas as pd
# Packages for plotting
import hvplot.pandas
import panel as pn
import matplotlib.pyplot as plt
# Useful for ignoring warnings that arise with gpd.read_file
import warnings
# Ignoring warnings is usually not recomended. However, all of the code
# is working and the warnings muddy up the exported HTML.
warnings.filterwarnings("ignore")
```

### Step 1: Wrangling Data

We need to get access to all of the data needed in order to run this analysis. For that, we can rely on the following scripts to access the drought data. Unfortunately, I was not able to find a way to do the same for the MTBS fire data. To download the data you need, navigate to https://www.mtbs.gov. Next, navigate to the data download site. Then, select California and set the date range to (2020-2020). Once you have that, you should be able to view all of the fires for the state and year. Select all of the files, input your email, and click Download Data. Once you have done that, MTBS will email you a link to download firepackages for each fire in your selection. Download this data, and move it into the fire data directory that will be created in the code below. 

*Note*: The fire data comes zipped and nested in some folders. You should extract the data, and remove the nested structure. This should leave you with a folder containing zipfiles. Lastly, rename your folder to something human readable. For me, this is simply "fire".

```python
# Define and create all filepaths
PROJECT_DIRECTORY = pathlib.Path.cwd().parent
DATA_DIRECTORY = PROJECT_DIRECTORY / "data"
DROUGHT_DIRECTORY = DATA_DIRECTORY / "drought"
FIRE_DIRECTORY = DATA_DIRECTORY / "fire"
os.makedirs(DATA_DIRECTORY, exist_ok=True)
os.makedirs(DROUGHT_DIRECTORY, exist_ok=True)
```

```python
# Download drought data
DROUGHT_URL = "https://droughtmonitor.unl.edu/data/shapefiles_m//2020_USDM_M.zip"
DROUGHT_FILE = DROUGHT_URL.split("/")[-1]
DROUGHT_FILEPATH = os.path.join(DATA_DIRECTORY, DROUGHT_FILE)

# Check to see if data is already downloaded.
# If not, send request to URL and download data from there.
if os.path.exists(DROUGHT_FILEPATH):
    print("Alreadying downloaded drought data, skipping...")
else:
    with requests.get(DROUGHT_URL, stream=True) as r:
        r.raise_for_status()
        with open(DROUGHT_FILEPATH, "wb") as f:
            for chunk in r.iter_content(chunk_size=8192):
                f.write(chunk)
# Extract zip file
with ZipFile(DROUGHT_FILEPATH, 'r') as zipObject:
    zipObject.extractall(DROUGHT_DIRECTORY)
# Delete old zip file
os.remove(DROUGHT_FILEPATH)
```

Now that we have all of the data, we need to wrangle it into a workable format. 

```python
# Get california bounds so we can clip the drought shapefiles to only california.
california_gdf = osm.geocode_to_gdf("California").to_crs(crs="EPSG:4326")
```

```python
# Open all fire shapefiles
fire_files = sorted(list(FIRE_DIRECTORY.glob("*.zip")))
fire_gdfs = []
for file in fire_files:
    path = FIRE_DIRECTORY / file
    bounds = gpd.read_file(path).to_crs(crs="EPSG:4326")
    date = bounds.Event_ID[0][-8:] # Get the date from the filename
    date = datetime.datetime.strptime(date, "%Y%m%d")
    bounds['week'] = date.isocalendar()[1]
    fire_gdfs.append(bounds)

# Merge all the GDFs into one
merged_fire_gdf = gpd.GeoDataFrame(pd.concat(fire_gdfs, ignore_index=True))
# Select only columns we want
merged_fire_gdf = merged_fire_gdf[['Incid_Name', 'BurnBndAc', 'geometry', 'week']]
merged_fire_gdf.head()
```

```python
# Open all drought shapefiles
# Working with drought shapefiles
drought_files = sorted(list(DROUGHT_DIRECTORY.glob("*.zip")))
drought_gdfs = []
for file in drought_files:
    bounds = gpd.read_file(file).to_crs(crs="EPSG:4326")
    bounds = gpd.clip(bounds,california_gdf) # Clip drought to bounds of California
    date = str(file)[-14:-6] # Extract date from filename
    date = datetime.datetime.strptime(date, "%Y%m%d")
    bounds['week'] = date.isocalendar()[1] - 1 
    drought_gdfs.append(bounds)
# Merge into one GDF
merged_drought_gdf = gpd.GeoDataFrame(pd.concat(drought_gdfs, ignore_index=True))
merged_drought_gdf = merged_drought_gdf[['DM', 'geometry','week']]
merged_drought_gdf.head()
```

### Step 2: Analysis
In order to analyze the relationship between drought and wildfires, I am going to look at the number of fires started in drought conditions and the total acres of those fires. While this doesn't give a perfect picture of the relationship, it does give us some cool insights.

```python
# Create a different GDF of just fire points for more efficient plotting and counting
fire_points = merged_fire_gdf.copy()
fire_points['geometry'] = fire_points['geometry'].centroid
```

```python
# Counting the points that are within drought vs not within drought
fires_in_drought = (
    merged_drought_gdf
    .to_crs(fire_points.crs)
    .sjoin(
        fire_points,
        how='inner',
        predicate='contains'
    )
).query('week_left == week_right')

# Defining number of fires and acres burned inside and outside of drought conditions
num_inside_fire = len(fires_in_drought)
num_outside_fire = len(merged_fire_gdf) - num_inside_fire
total_acres_burned = merged_fire_gdf['BurnBndAc'].sum()
burned_acres_inside = fires_in_drought['BurnBndAc'].sum()
burned_acres_outside = total_acres_burned - burned_acres_inside
print(f'Number of fires ignited in areas with drought conditions: {num_inside_fire}')
print(f'Number of fires ignited outside of areas with drought conditions: {num_outside_fire}')
print(f'Total number of acres burned: {total_acres_burned}')
print(f'Total acres burned by fires ignited within drought conditions: {burned_acres_inside}')
print(f'Total acres burned by fires ignited outside of areas with drought conditions: {burned_acres_inside}')
print(f'% of acres burned in 2020 by fires ignited within drought conditions: {(burned_acres_inside/total_acres_burned)*100}%')
```

### Step 3: Plotting/Analysis
Now that we have all of the data we want to look at, we should create some plots to analyze what is going on.

```python
# Defining our static plots

# Define lists for the fire counts
count_labels = ['Outside','Inside']
counts = [num_outside_fire, num_inside_fire]
# Define plots
fig_count, ax_count = plt.subplots()
# Set labels, data to be plotted, etc.
ax_count.bar(count_labels,counts)
ax_count.set_title('Number of Fires Started \n in Drought Area vs. Outside Drought Area \n California 2020')
ax_count.set_ylabel('Number of Fires')

# Define lists for the burned acres
acres_labels = ['Acres Outside', 'Acres Inside', 'Total Acres']
acres = [burned_acres_outside, burned_acres_inside, total_acres_burned]
#Define plots
fig_acre, ax_acre = plt.subplots()
# Set labels, data to be plotted, etc.
ax_acre.bar(acres_labels,acres)
ax_acre.set_title('Comparing Total Acres Burned for \n Fires Started Inside vs. Outside a Drought Area \n California 2020')
ax_acre.set_ylabel('Acres Burned (Millions)')
```

As we can tell from the bar charts above, we see that ignition events happened more frequently in areas experience drought or near drought conditions. We also see that the total acres burned by fires ignited within areas experiencing drought is a very large portion of the total acres burned in 2020.

Lets create some dynamic plots to better visualize what is going on.

```python
# Define our dynamic plots

# Set bounds so plot zooms to proper bounds
xmin, ymin, xmax, ymax = merged_drought_gdf.total_bounds

# Define custom colormap for viewing the drought areas
drought_colormap = {0:'#FFFF00', 1:'#FCD37F', 2:'#FFAA00', 3:'#E60000', 4:'#730000'}
# Create fill color variable for plotting
merged_drought_gdf['fill_color'] = merged_drought_gdf['DM'].map(drought_colormap)

fire_points_plot = fire_points.hvplot(geo=True, tiles="CartoLight",
                                  fill_color="red", 
                                  frame_width=600, frame_height=600,
                                  groupby="week", # Groupby week to enable slider
                                  xlim=(xmin,xmax),ylim=(ymin,ymax), # Set view bounds
                                  title="2020 California Drought and Fires",
                                  size=100
                                  )


drought_plot = merged_drought_gdf.hvplot(geo=True, tiles="CartoLight",
                                        color='fill_color', # Fill by the fill color
                                        frame_width=600, frame_height=600,
                                        groupby="week", fill_alpha=1, # Groupby week to enable slider
                                        xlim=(xmin,xmax),ylim=(ymin,ymax), # Set view bounds
                                        title="2020 California Drought and Fires"
                                        )

fire_bounds_plot = merged_fire_gdf.hvplot(geo=True, tiles="CartoLight",
                                         fill_color="black",
                                         frame_width=600, frame_height=600,
                                         groupby="week", # Groupby week to enable slider
                                         xlim=(xmin,xmax), ylim=(ymin,ymax), # Set view bounds
                                         title="2020 California Drought and Fires",
                                         )
```

```python
# Show the points plot
points_plot = (fire_points_plot * drought_plot)
points_plot
```

I was not able to get a legend working for the above plot. Just note that the range corresponds with the U.S. Drought Monitor indices. The colors are taken directly from their color scheme. You can read more about the indices [here](https://www.drought.gov/data-maps-tools/us-drought-monitor). Just note the lighter darker colors are considered more extreme drought conditions than the lighter colors.

As we can see, the drought shapefiles line up fairly well with the report published on the 2020 drought season [[3]](https://www.drought.gov/drought-status-updates/drought-status-update-and-2020-recap-california-nevada). We see our first fire ignited late april/early May. This fire is just barely within an area experiencing abnormally dry conditions. However, it and the surrounding area had been experiencing drought or dry conditions since Febuary.

The most interesting week to look at is week 34. Here we see a total of 20 fires ignited, with only 3 of them being ignited in areas not experiencing drought. While this is interesting and confirms the idea that drought makes the ignition of fires more likely, looking at the area burned starting this week is even more telling.

```python
# Show the fire bounds plot
bounds_plot = (drought_plot * fire_bounds_plot)
bounds_plot
```

Be patient when scrolling to the 34th week on this plot. The boundaries take a long time to render due to the large number of bounds, and the complexity of those bounds.

Here we see some very large fires were ignited this week. The largest fire on California record was ignited, the August Complex. The August complex started as a bunch of smaller fires ignited by a lightning storm that eventually expanded into each other creating one massive complex. This fire burned over 1 million acres during its 87 day lifespan [[4]](https://www.fire.ca.gov/incidents/2020/8/16/august-complex-includes-doe-fire). This amounts to nearly 1% of the total landmass of the state. As we can see from the plot, this area experienced drought for most of the year leading up to the ignition, and sustained it well past the fire being contained.

## Results/Discussion
From this analysis, we can see that drought did play a roll in the 2020 California fire season. We saw 67 of the 93 fires, roughly 72%, ignited during this year ignited in areas that were experiencing abnormally dry conditions or drought. Furthermore, we saw a wopping 88% of the total acres burned during this period burned by fires ignited within drought conditions.

While this data points to the conclusion of drought worsening fire season, there is likely more than just that going on here. Previous research has shown that La Nia creates atmospheric conditions that increase the intensity of fire seasons [[1]](https://doi.org/10.1038/s41612-022-00248-4). According to the [Golden Gate Weather Services](https://ggweather.com/enso/oni.htm), ENSO began trending towards La Nia starting and peaking in 2020. From this, we can see a correlation between the intensity/severity of fires and the onset of La Nia. This is hinting back at the compound effect of drought and atmospheric events that create worse fire weather [[1]](https://doi.org/10.1038/s41612-022-00248-4).

I would lastly like to talk about the conclusions that can be drawn from this type of analysis. Because my analysis was limited to looking at the initial date of ignition events and drought boundaries, there are many confounding variables that are not being taken into account. For example, stronger winds may have been prevelent during the worse weeks of this fire year, thus causing more spread. Also, limiting the data to 1 year that is known to have been the worst fire year on record, and consistent drought conditions could be pointing me towards a conclusion that I want to see. A longer timespan of data is needed in order to draw stronger conclusions. Lastly, the study is isolated to California. While California does have a wide variety of terrain and ecosystems, we cannot make generalizations about the entire world based on studies isolated to one region.

With all of that being said, this analysis shows that there was atleast some affect from drought on the 2020 California fire season.
