```python
import hvplot.pandas
import panel as pn
import pandas as pd
import geopandas as gpd
```

```python
# Creating plots
xmin, ymin, xmax, ymax = merged_drought_gdf.total_bounds

drought_colormap = {0:'#FFFF00', 1:'#FCD37F', 2:'#FFAA00', 3:'#E60000', 4:'#730000'}
merged_drought_gdf['fill_color'] = merged_drought_gdf['DM'].map(drought_colormap)

slider_widget = pn.widgets.IntSlider(
    name="Week of Year",
    start=1,
    end=52,
    step=1,
    value=1,
    width=450
)

fire_plot = merged_fire_gdf.hvplot(geo=True, tiles="CartoLight",
                                  fill_alpha=0.5, line_width=2,
                                  line_color="black", fill_color="red",
                                  frame_width=600, frame_height=600,
                                  groupby="week",
                                  xlim=(xmin,xmax),ylim=(ymin,ymax),
                                  #widget_location="bottom", widgets={"Week of Year": slider_widget},
                                  title="2020 California Drought and Fires"
                                  )


drought_plot = merged_drought_gdf.hvplot(geo=True, tiles="CartoLight",
                                        color='fill_color',
                                        frame_width=600, frame_height=600,
                                        groupby="week",
                                        xlim=(xmin,xmax),ylim=(ymin,ymax),
                                        #widget_location='bottom',
                                        #clim=(0,4), widgets={"Week of Year": slider_widget},
                                        title="2020 California Drought and Fires"
                                        )
```

```python
full_plot = drought_plot * fire_plot
dashboard = pn.Column(
    full_plot
)
dashboard
```
