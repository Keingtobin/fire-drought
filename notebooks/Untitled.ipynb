```python
# Imports
import os
import requests
import pathlib
import geo
```

```python
# Download drought data
DROUGHT_URL = "https://droughtmonitor.unl.edu/data/shapefiles_m//2020_USDM_M.zip"
DROUGHT_FILE = DROUGHT_URL.split("/")[-1]
PROJECT_DIRECTORY = pathlib.Path.cwd().parent
DATA_DIRECTORY = PROJECT_DIRECTORY / "data"
DROUGHT_DIRECTORY = DATA_DIRECTORY / "drought"
os.makedirs(DATA_DIRECTORY, exist_ok=True)
os.makedirs(DROUGHT_DIRECTORY, exist_ok=True)
```

```python
DROUGHT_FILEPATH = os.path.join(DROUGHT_DIRECTORY, DROUGHT_FILE)
if os.path.exists(DROUGHT_FILEPATH):
    print("Alreadying downloaded drought data, skipping...")
else:
    with requests.get(DROUGHT_URL, stream=True) as r:
        r.raise_for_status()
        with open(DROUGHT_FILEPATH, "wb") as f:
            for chunk in r.iter_content(chunk_size=8192):
                f.write(chunk)
```
